```python
# ลบแถวที่ไม่ต้องการใน DataFrame ด้วย drop()
df.drop([0, 1])

df.drop([2,3,4]).head()
```

```python
# การแทนที่ข้อมูลใน DaaFrame ด้วย Pandas
# df.replace({'column_name': {'search': 'replace'}}).head(10)
df.replace({'Pclass': {1: '1st', 2:'2nd', 3:'3rd'}})

# การแทนที่ข้อมูลทั้งหมดใน DataFrame ด้วย Pandas
df.replace(0, 'Zero')
# df.replace('NaN', 0)
```

```python
# ตัวอย่างการนับข้อมูลเป็นกลุ่ม
# เช่นนับแยกตั๋วโดยสารแต่ละชั้น
# เรียงลำดับ index จากน้อยไปมาก
df['Pclass'].value_counts().sort_index()
```

#### ทำความสะอาดทุกเงื่อนไข

```python
df = pd.read_csv('dataset/exercise.csv')

# แทนที่ข้อมูลว่างด้วยคำสั่ง fillna
df.fillna(-99,inplace=True)

# เปลี่ยนข้อมูลที่ผิดให้ถูกต้อง
df.loc[2, 'Hight'] = 179

# ลบรายการที่ซ้ำกันทั้งหมดออก
df.drop_duplicates(inplace=True)

# ลบแถวที่มีรายการผิดรูปแบบ
df.drop([3])
```
