#### Step 3: วาดกราฟแท่ง (Bar Chart)

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# กำหนดแกน x และ y
# x = np.array([2018, 2019, 2020, 2021, 2022])
# y = np.array([1120, 2544, 3588, 4215, 4052])

df = pd.read_csv('datasets/gdp.csv')

plt.bar(df['year'], df['gdp'], color=['green','red','blue','yellow','pink'], edgecolor='black')
plt.title('Thailand GDP')
plt.xlabel('Year')
plt.ylabel('GDP (Million Baht)')
plt.show()
```
